```java
// 2019 Question 1 - (8/9)

public static int numberOfLeapYears(int year1, int year2) {
    int res = 0; // +1
    for (int i = year1; i <= year2; i++) { // +1
        if (isLeapYear(i)) { // +1
            res++; // +1
        }
    }
    return res; // +1
}

public static int dayOfWeek(int month, int day, int year) {
    int startDay = firstDayOfYear(year); // +1
    int weekDaysPassed = (dayOfYear(month, day, year) % 7); // +1
    return startDay + weekDaysPassed; // +0: Week starts on day 0, so subtract by 1 necessary; +1
}

// 2019 Question 2 - (7/9)
public Class StepTracker {
    private int activeReq = 0; // +1
    private ArrayList<Integer> stepsList;
    StepTracker(int req) { // +0, forgot public, +1
        this.activeReq = req;
    }
    addDailySteps(int steps) { // +0, forgot public void, +1, +1
        this.stepsList = new ArrayList<Integer>()
        stepsList.add(steps);
    }
    public int activeDays() { // +1
        int ret = 0;
        for (int day : steps) {
            if (day >= activeReq) {
                ret++;
            }
        }
        return ret
    }
    public double averageSteps() { // +1, +1
        int sum = 0;
        for (int day : steps) {
            sum += day;
        }
        return (double) sum/steps.length()
    }
}

// 2019 Question 3 - (9/9)
public ArrayList<String> getDelimitersList(String[] tokens) {
    ArrayList<String> ret = new ArrayList<String>(); // +1
    for (String token : tokens) { // +1
        if (token.equals(openDel) || token.equals(closeDel)) { // +1
            ret.add(token) // +1
        }
    }
    return ret;
}

public boolean isBalanced(ArrayList<String> delimiters) {
    int open = 0; // +1
    int close = 0;
    for (String token : delimiters) { // +1
        if (token.equals(openDel)) { // +1
            open++;
        }
        if (token.equals(closeDel)) {
            close++;
        }
    }
    if (open == close) {
        return true; // +1
    }
    return false; // +1
}

// 2019 Question 4 - (9/9)
public LightBoard(int numRows, int numCols) {
    this.lights = new boolean[numRows][numCols]; // +1
    for (int i=0; i<numRows; i++) { // +1
        for (int j=0; j<numCols; j++) {
            lights[i][j] = (Math.random <= 0.4); // +1, +1
        }
    }
}

public boolean evaluateLight(int row, int col) {
    int lights = 0;
    if (lights[row][col]) { // +1
        for (int j=0; j<col; j++) { // +1
            if (lights[row][j]) { // +1
                lights++;
            }
        }
        if (lights % 2 == 0) { // +1
            return false; // +1
        }
    } else {
        lights = 0;
        for (int j=0; j<col; j++) {
            if (lights[row][j]) {
                lights++;
            }
        }
        if (lights % 3 == 0) {
            return true;
        }
    }
    return (lights[row][col]);
}
```
